## 7.1 전체 구조
- 합성곱 계층
- 풀링 계층

- 완전 연결(전결합) : 인접하는 계층의 모든 뉴런과 결합된 것
-> 완전히 연결된 계층을 Affine 계층이라는 이름으로 구현하였음
-> Affine 계층 뒤에 활성화 함수를 갖는 ReLU 계층이 이어짐 (Affine-ReLU 조합), 마지막 계층은 Affine-SoftMax 조합

- CNN에서는 Conv-ReLU-(Pooling) 조합
-> 출력에 가까운 층에서는 Affine-ReLU 구성 사용 가능, 마지막 계층은 Affine-SoftMax 조합 그대로 사용


## 7.2 합성곱 계층
- 패딩, 스트라이드
- 완전 연결 계층의 문제점 : 데이터의 형상이 무시됨
-> 완전 연결 계층에 입력할 때에는 3차원 데이터를 1차원 데이터로 평탄화해서 전달해야하기 때문

CNN 에서는 합성곱 계층의 입출력 데이터를 특징맵(feature map)이라 함
(입출력데이터 = 특징맵)

입력 데이터에 필터(=커널)을 적용함, 이때 합성곱 연산을 수행함

합성곱 연산 : 필터의 윈도우를 일정 간격 이동해가며 입력 데이터에 적용함
-> 입력과 필터에서 대응하는 원소끼리 곱한 후 그 총합을 구함 (이를 FMA : 단일곱셈-누산 이라 함)

패딩 : 입력데이터 주변을 특정 값(ex. 0)으로 채우기도 함

스트라이드 : 기본적 합성곱은 스트라이드가 1이다
-> 스트라이드가 2가 되면 필터를 적용하는 윈도우가 2칸씩 이동함

3차원 합성곱을 곱할 때에는 각 입력데이터별로 필터와 합성곱한 뒤 모든 채널의 값을 합한다.
-> 주의점 : 입력 데이터의 채널 수와 필터의 채널 수가 같아야 함

블록으로 생각할 경우, 합성곱 연산의 출력으로 다수의 채널을 내보내려면 필터(가중치)를 다수 사용하는 것
-> 편향을 추가할 수 있음(완전 연결 계층과 마찬가지로)

배치 처리 : 신경망에 4차원 데이터가 흐르면 데이터 n개에 대한 합성곱 연산이 이뤄짐 -> n회분의 처리를 한 번에 수행하는 것

## 7.3 풀링 계층
풀링 : 세로/가로 방향의 공간을 줄이는 연산, n X n 영역을 원소 하나로 집약하여 공간 크기를 줄임
최대 풀링 : 영역 내 최대값을 구하는 연산

풀링계층의 특징:
- 학습할 매개 변수가 없음
- 채널 수가 변하지 않음
- 입력의 변화에 영향을 적게 받는다
